```python
# Traffic Accident Data Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Plot Style
sns.set_style("whitegrid")
# Create Sample Dataset
data = {
    "Weather": [
        "Clear", "Rain", "Fog", "Clear",
        "Rain", "Fog", "Clear", "Rain"
    ],
    
    "Road_Condition": [
        "Dry", "Wet", "Wet", "Dry",
        "Wet", "Wet", "Dry", "Wet"
    ],
    
    "Time_of_Day": [
        "Morning", "Night", "Evening", "Afternoon",
        "Night", "Morning", "Evening", "Night"
    ],
    
    "Severity": [2, 4, 3, 1, 4, 3, 2, 5]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Dataset

print("\n===== Traffic Accident Dataset =====\n")
print(df)

# Create Multiple Graphs Together
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Accidents by Weather Condition

sns.countplot(
    x="Weather",
    data=df,
    palette="Blues",
    ax=axes[0, 0]
)

axes[0, 0].set_title("Accidents by Weather")
axes[0, 0].set_xlabel("Weather Condition")
axes[0, 0].set_ylabel("Count")

# 2. Accidents by Road Condition

sns.countplot(
    x="Road_Condition",
    data=df,
    palette="Greens",
    ax=axes[0, 1]
)

axes[0, 1].set_title("Accidents by Road Condition")
axes[0, 1].set_xlabel("Road Condition")
axes[0, 1].set_ylabel("Count")

# 3. Accidents by Time of Day

sns.countplot(
    x="Time_of_Day",
    data=df,
    palette="Oranges",
    ax=axes[1, 0]
)

axes[1, 0].set_title("Accidents by Time of Day")
axes[1, 0].set_xlabel("Time")
axes[1, 0].set_ylabel("Count")

# Rotate labels for better visibility
axes[1, 0].tick_params(axis='x', rotation=15)

# 4. Severity Distribution

sns.histplot(
    df["Severity"],
    bins=5,
    kde=True,
    color="red",
    ax=axes[1, 1]
)

axes[1, 1].set_title("Accident Severity Distribution")
axes[1, 1].set_xlabel("Severity")
axes[1, 1].set_ylabel("Frequency")

# Adjust Layout
plt.tight_layout()

# Show All Plots
plt.show()

# Accident Hotspot Visualization

plt.figure(figsize=(8, 5))

x_locations = [1, 2, 3, 4, 5, 6, 7, 8]
y_locations = [2, 3, 1, 5, 4, 6, 5, 7]

sns.scatterplot(
    x=x_locations,
    y=y_locations,
    hue=df["Severity"],
    size=df["Severity"],
    palette="coolwarm",
    sizes=(100, 400)
)

plt.title("Accident Hotspot Visualization")
plt.xlabel("Location X")
plt.ylabel("Location Y")

plt.grid(True)
plt.show()
```
